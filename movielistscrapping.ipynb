```python
import requests
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm
```

```python
#https://en.wikipedia.org/wiki/Lists_of_Tamil-language_films
```

```python
url = "https://en.wikipedia.org/wiki/Lists_of_Tamil-language_films"
```

```python
response = requests.get(url)
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
#soup
```

```python
paradigms_div = soup.find("div", {"id": "bodyContent"})
```

```python
paradigms_list_items = paradigms_div.find_all("a")
```

```python
paradigms_list_items
comparison_lists = ['https://en.wikipedia.org'+lst['href'] for lst in paradigms_list_items if lst.has_attr("title") and lst["title"].startswith("List of Tamil films") and lst.text.startswith("List of Tamil films of")][1:]
```

```python
comparison_lists
```

```python
#https://en.wikipedia.org/wiki/List_of_Tamil_films_of_1940
#https://en.wikipedia.org/wiki/List_of_Tamil_films_of_2005
```

```python
movie_list = []
```

```python
for movie_url in tqdm(comparison_lists):
  url1 = movie_url
  response1 = requests.get(url1)
  html1 = response1.content
  soup1 = BeautifulSoup(html1, "html.parser")
  tb = soup1.find_all("table",{'class':"wikitable"})
  for j in tb:
    tf = j.find_all("tr")
    for k in tf:
      cv = k.find_all("td")
      for kd in cv:
        to = kd.find("a")
        if to is not None and to.has_attr('title') and to.has_attr('href'):
          movie_list.append([to['title'],to['href']])
          break
```

```python
len(movie_list)
```

```python
import csv
with open("moviename.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    for row in movie_list:
        writer.writerow(row)
```
