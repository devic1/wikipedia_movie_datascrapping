```python
for index, row in tqdm(data.iterrows()):
  try:
    pl = row['Wiki Page']
    url = pl
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html, "html.parser")
    if response == 404:
      pass
    try:
      movbox = soup.find("table", {"class": "infobox vevent"})
      l = movbox.find("a",{'class':'image'})
      u = l.find("img")['src']
      image = "https:"+u
      data.loc[index, 'wiki_img'] = image
    except:
      inbox_una.append(url)
  except:
    print(row['Wiki Page'])
```

```python
for ind, ro in tqdm(data.iterrows()):
  try:
    li = ro['wiki_img']
    s = ""
    for q,u in enumerate(li):
      s += u
      if u == '.' and li[q+1] == 'j' and li[q+2] == 'p' and li[q+3] == 'g':
        s += 'jpg'
        break
    su = re.sub(r'/thumb', '',s)
    data.loc[ind,'r_img'] = su
  except:
    print(ro)
```
